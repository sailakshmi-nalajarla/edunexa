```python
dia_df['Glucose'].fillna(dia_df['Glucose'].mean(),inplace=True)
dia_df['BloodPressure'].fillna(dia_df['BloodPressure'].mean(),inplace=True)
dia_df['SkinThickness'].fillna(dia_df['SkinThickness'].median(),inplace=True)
dia_df['Insulin'].fillna(dia_df['Insulin'].median(),inplace=True)
dia_df['BMI'].fillna(dia_df['BMI'].median(),inplace=True)
dia_df.isnull().sum()
```

```python
#EDA
#Bar chart
plt.figure(figsize=(10,5))
plt.title('Diabetes Plot Yes/No', fontsize=14)
sns.countplot(x="Outcome", data=dia_df, palette=('#FFFF00','#800080'))
plt.xlabel("Diabetes (0 = No, 1= Yes)", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
fig, axes = plt.subplots(2, 4, figsize=(18, 12))
fig.suptitle('Diabetes Outcome Distribution WRT All Independent Variables', fontsize=16)

sns.boxplot(ax=axes[0, 0], x=dia_df['Outcome'], y=dia_df['Pregnancies'],
            hue=dia_df['Outcome'], palette=('#FFFF00', '#800080'))
axes[0, 0].set_title("Diabetes Outcome vs Pregnancies", fontsize=12)

sns.boxplot(ax=axes[0, 1], x=dia_df['Outcome'], y=dia_df['Glucose'],
            hue=dia_df['Outcome'], palette=('#FFFF00', '#800080'))
axes[0, 1].set_title("Diabetes Outcome vs Glucose", fontsize=12)

sns.boxplot(ax=axes[0, 2], x=dia_df['Outcome'], y=dia_df['BloodPressure'],
            hue=dia_df['Outcome'], palette=('#FFFF00', '#800080'))
axes[0, 2].set_title("Diabetes Outcome vs BloodPressure", fontsize=12)

sns.boxplot(ax=axes[0, 3], x=dia_df['Outcome'], y=dia_df['SkinThickness'],
            hue=dia_df['Outcome'], palette=('#FFFF00', '#800080'))
axes[0, 3].set_title("Diabetes Outcome vs SkinThickness", fontsize=12)

sns.boxplot(ax=axes[1, 0], x=dia_df['Outcome'], y=dia_df['Insulin'],
            hue=dia_df['Outcome'], palette=('#FFFF00', '#800080'))
axes[1, 0].set_title("Diabetes Outcome vs Insulin", fontsize=12)

sns.boxplot(ax=axes[1, 1], x=dia_df['Outcome'], y=dia_df['BMI'],
            hue=dia_df['Outcome'], palette=('#FFFF00', '#800080'))
axes[1, 1].set_title("Diabetes Outcome vs BMI", fontsize=12)

sns.boxplot(ax=axes[1, 2], x=dia_df['Outcome'], y=dia_df['DiabetesPedigreeFunction'],
            hue=dia_df['Outcome'], palette=('#FFFF00', '#800080'))
axes[1, 2].set_title("Diabetes Outcome vs DiabetesPedigreeFunction", fontsize=12)

sns.boxplot(ax=axes[1, 3], x=dia_df['Outcome'], y=dia_df['Age'],
            hue=dia_df['Outcome'], palette=('#FFFF00', '#800080'))
axes[1, 3].set_title("Diabetes Outcome vs Age", fontsize=12)
```

```python
from sklearn.model_selection import train_test_split
x = dia_df.drop(['Outcome'],axis=1)
y = dia_df['Outcome']
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x_scaled= sc.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.3, random_state=0)
x_train.shape, y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
y_pred = logreg.predict(x_test)
#confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score
confmat = confusion_matrix(y_pred, y_test)
confmat

from sklearn import metrics
cm = metrics.ConfusionMatrixDisplay(confusion_matrix=metrics.confusion_matrix(y_pred, y_test, labels=logreg.classes_),
display_labels=logreg.classes_)
cm.plot(cmap="PuBuGn")
```

```python
accuracy_score(y_pred, y_test)
```
